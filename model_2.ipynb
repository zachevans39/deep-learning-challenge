```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf


#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv('../Starter_Code/Resources/charity_data.csv')
application_df.head()
```

```python
new_app=application_df.drop(['EIN','NAME'],axis=1)
new_app
```

```python
applications= pd.get_dummies(new_app,dtype=float)
applications
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y= applications['IS_SUCCESSFUL'].values
print(y)

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X=applications.drop('IS_SUCCESSFUL',axis=1).values
print (X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
number_features = len(X_train_scaled[0])
nodes_layer1=80
nodes_layer2=50

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=nodes_layer1, input_dim=number_features, activation='relu'))
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=nodes_layer2,activation='relu'))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE


fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
